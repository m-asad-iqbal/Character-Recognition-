# Character Recognition from Own Captured Image

# Importing Already Trained Model

```python
# Importing Joblib Library
import joblib 
# Importing model that is already trained on Hand Recognition Problem
model=joblib.load('Digit-Recognition-model')
model
```

# Importing Image

```python
from PIL import Image
import matplotlib.pyplot as plt
# Importing Image that is taken from mobile camera (containing a character)
image =Image.open('six.png')
plt.imshow(image)
```

# Converted RGB to GrayScale

```python
import numpy as np
from PIL import Image, ImageOps
im2 = ImageOps.grayscale(image)
plt.imshow(im2, cmap=plt.cm.gray)

# image = image.convert("L")
# plt.imshow(image)
```

# Resizing the GrayScale Image to 8 x 8 

```python
image = image.resize((8, 8), Image.ANTIALIAS)
plt.imshow(image)
```

# Convert New Image to Matrix Form

```python
matrix = np.array(image)*(16/255)
print(matrix)
```

# Flatten the Matrix into 1D array in order to feed in the model

```python
# Flatten the matrix into a 1D array
flattened_matrix = matrix.flatten()
```

```python
flattened_matrix.shape
```

# Predicting the Character in the Image

```python
image_prediction = model.predict([flattened_matrix])

print("Predicted character:", image_prediction[0])
```
